```python
import pandas as pd
import numpy as np
import nltk
from google.colab import drive
from nltk.tokenize import RegexpTokenizer
# KERAS
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
drive.mount('/content/gdrive')
```

```python
df_processed = pd.read_csv('gdrive/MyDrive/save_file/20220218/data_test_processed_formal.csv')
```

```python
tokenizer_data = RegexpTokenizer(r'\w+')
df_processed['value_tokenize'] = df_processed['review_formal_processed'].map(tokenizer_data.tokenize)
```

```python
value = df_processed["value_tokenize"]
```

```python
import pickle

with open('gdrive/MyDrive/save_file/20220218/tokenizer.pickle', 'rb') as handle:
    loaded_tokenizer = pickle.load(handle)
```

```python
from keras.models import load_model

# load model from single file
model = load_model('gdrive/MyDrive/save_file/20220218/lstm.h5')
```

```python
seq = loaded_tokenizer.texts_to_sequences(value)
padded = pad_sequences(seq, padding = 'post')
pred = model.predict(padded)
print(pred)
```

```python
np.savetxt('pred_test_20220219.csv',pred, delimiter=',')
```

```python
act_labels = {0:0, 1:1, 2:2, 3:3}
```

```python
act_labels
```

```python
rev_act_labels = {val: key for key, val in act_labels.items()}
predictions = np.argmax(pred, axis=1)
act_predictions  = [rev_act_labels[val] for val in predictions]
print(act_predictions)
```

```python
act_predictions
```

```python
ser = pd.Series(act_predictions)
```
